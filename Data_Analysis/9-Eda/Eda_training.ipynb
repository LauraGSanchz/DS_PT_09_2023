## Cardinalidad ##

% de cardinalidad Dividimos en 3:
- cuantos valores unicos hay
- relacion valores unnicos respecto filas qeu hay- 
- tipo de dato 

manualmente hay qeu dividir entre tipos de datos, decidiendo segun lo que informa la variable. tiene que quedar algo cmoo lo de abajo.



```python
for i in df.select_dtypes(include=object):
    print(f"La cardinalidad de la variable {i} es {(df[i].nunique()/df.shape[0])*100:.2f}%")
```
