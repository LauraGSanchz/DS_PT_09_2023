## **Crear DataFrame** atacando la `API`

Hay que crear una función llamada hash_params():

```python
def hash_params(timestamp,priv_key,pub_key):
    """Marvel API requieres server side API calls to include 
    md5 hash of timestamp + public key + private key"""
    
    hash_md5 = hashlib.md5()
    hash_md5.update(f'{timestamp}{priv_key}{pub_key}'.encode('utf-8'))
    hashed_params = hash_md5.hexdigest()

    return hashed_params
```

```python
# 1 obtener lista de todos los superheroes que empiezan por L
timestamp = datetime.datetime.now().strftime("%Y/%m/%d%H:%M:%S")

pub_key = 'c61903313ea2c85d301a62dbf987f352'
priv_key = '7cffcb8c2d854f1f6adcea3d274c8e2a79f5ac0b'

params = {'limit': 100,
          'nameStartsWith': 'l',
          'ts': timestamp,
          'apikey': pub_key,
          'hash' : hash_params(timestamp,priv_key,pub_key)}

url = 'http://gateway.marvel.com/v1/public/characters'

res = requests.get(url,params)
res.json()
```

```python
res.status_code
```

Necesitamos que el DataFrame tenga la siguiente informacion:
1. id
2. nombre
3. url imagen
4. comics donde aparece
5. creador


```python
data = res.json()
```

```python
results = data['data']["results"]
results
```

```python
marvel_list = [{'id': hero['id'],'name': hero['name'],'thumbnail': hero['thumbnail']} for hero in data['data']["results"] ]

heroes =pd.DataFrame(marvel_list)
heroes.set_index('name', inplace=True)
```

```python
dict_comics = {}
for personaje in results:
    nombre_personaje = personaje['name']
    comics_personaje = [comic['name'] for comic in personaje['comics']['items']]
    dict_comics[nombre_personaje] = comics_personaje
```

```python
new_column= pd.Series(dict_comics)
```

```python
heroes['comics'] = new_column
heroes
```
