### Paso 4. Descubre cuál es la edad media por ocupación

```python
users.groupby('occupation')[['age']].mean()

```

### Paso 5. Descubra la proporción de Hombres por ocupación y ordénela de mayor a menor

#### Forma 1

```python
proporcion = users.groupby('occupation')['gender'].apply(lambda x: (x == 'M').sum() / len(x))
proporcion.sort_values()
```

#### Forma 2

```python
proporcion_genero = users.pivot_table(index='occupation', 
                                   columns='gender', 
                                   aggfunc='size', 
                                   fill_value=0
                                   )

proporcion_genero
```

```python
proporcion_genero['proporcion_M'] = proporcion_genero['M'] / (proporcion_genero['M'] + proporcion_genero['F'])

prop_sorted = proporcion_genero['proporcion_M'].sort_values()
print(prop_sorted)
```

#### Forma 3

```python
prop_3 = users.groupby('occupation')['gender'].value_counts(normalize=True).unstack()
prop_3
```

```python
prop3_sorted = prop_3['M'].sort_values()

prop3_sorted
```
