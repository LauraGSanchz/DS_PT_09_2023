# Nombre del producto de valor máximo

```python
chipo['item_price'].max()
# no está bien lo anterior, porque el return tiene un espacio, es un tipo object y el numero mas alto que puede conseguir es el 9, 
# el 16 empieza por 1, por loq ue para pd es menor qeu eel 9.

```

```python
chipo['numero'] = chipo['item_price'].str[1:].str.rstrip()
```

```python
chipo.info()
# numero sigue siendo object
```

```python
chipo['numero'] = chipo['numero'].astype('float')
# ahora sí puedo hacer el maximo

chipo['numero'].max()
```

```python
# pero me pide el nombre, no el valor:
mask = chipo['numero'] == chipo['numero'].max()
chipo[mask]
# chipo[chipo['numero'] == chipo['numero'].max()]  esto es lo mismo que lo anterior.
```

```python
# pero lo anterior es el mayor precio pero de 15 unidades. Necesitamos una nueva columna de precio unitario
chipo['precio_unit'] = chipo['numero'] / chipo['quantity']
chipo['precio_unit'].max()
chipo[chipo['numero']== chipo['precio_unit'].max()].drop_duplicates(['item_name'])
```
